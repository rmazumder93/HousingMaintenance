```python
df.insert(loc=0, column='NovDimID', value=df.set_index(['ViolationID','NOVID','NOVDescription']).index.factorize()[0]+10)
df['NovDimID']
```

```python
df.insert(loc=0, column='AddressDimID', value=df.set_index(['ViolationID','BuildingID','BoroID','Borough','HouseNumber','StreetName','StreetCode','Postcode',
                                                            'Apartment','Story','Latitude','Longitude','BIN']).index.factorize()[0]+100)
df['AddressDimID']
```

```python
df.insert(loc=0, column='RegistrationDimID', value=df.set_index(['ViolationID','RegistrationID','OrderNumber']).index.factorize()[0]+1000)
df['RegistrationDimID']
```

```python
df.insert(loc=0, column='StatusDimID', value=df.set_index(['ViolationID','CurrentStatus','Class','CurrentStatus','CurrentStatusDate','ViolationStatus','RentImpairing']).index.factorize()[0]+10000)
df['StatusDimID']
```

```python
df.insert(loc=0, column='OrderID', value=df.set_index(['ViolationID','StatusDimID','RegistrationDimID']).index.factorize()[0]+20)
df['OrderID']
```
