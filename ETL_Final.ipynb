## Importing Data and Library

```python
import pandas as pd
df = pd.read_csv('Housing_Maintenance_Code_Violations.csv', sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
```

## About Datasets

```python
df.iloc[:20,:]
```

## Removing the following rows 

'CouncilDistrict','CommunityBoard','CensusTract','NTA','Block','Lot','BBL','HighHouseNumber','LowHouseNumber','OriginalCertifyByDate','OriginalCorrectByDate', 'NewCertifyByDate', 'NewCorrectByDate'

```python
remove = ['CouncilDistrict', 'CommunityBoard', 'CensusTract', 'NTA','Block','Lot','BBL','HighHouseNumber','LowHouseNumber','OriginalCertifyByDate',
       'OriginalCorrectByDate', 'NewCertifyByDate', 'NewCorrectByDate']

df = df.drop(columns=remove)
```

```python
print('Shape:', df.shape)
print('Column Names', df.columns.values)
```

```python
df.iloc[:20,:]
```

## Filter by order numbers 

Lead paint → 555, 606, 607, 610, 611, 612, 614, 616, 617
Mold → 550
Fire Escape → 181, 512, 513, 514, 515, 516, 517, 518, 519, 520, 539, 812, 813, 814, 815, 817, 818, 819, 820, 821
Heat → 226, 664, 665, 666, 668, 675, 964, 965, 975

```python
order_list = ['555','606','607','610','611','612','614','616','617','550','171','512','513','514','515',
              '516','517','518','519','520','539','812','813','814','815','817','818','819','820','821','226','664','665','666',
             '668','675','964','965','965']
filtered_df = df[df['OrderNumber'].isin(order_list)]
```

```python
print('Shape:', filtered_df.shape)
print('Column Names', filtered_df.columns.values)
filtered_df.iloc[:20,:]
```

## Filter by date [2020 - 2021]

```python
filtered_df['ApprovedDate'] = pd.to_datetime(filtered_df['ApprovedDate'], errors='coerce')
new_df = filtered_df[(filtered_df['ApprovedDate'] > '2020-01-01') & (filtered_df['ApprovedDate'] < '2021-12-31')]
```

```python
print('Shape:', new_df.shape)
print('Column Names', new_df.columns.values)
new_df.iloc[:20,:]
```

## Creating new row with violation type

```python
new_df["Violation Type"] = None
```

```python
new_df.shape
```

```python
lead_paint = ['555', '606', '607', '610', '611', '612', '614', '616', '617']
mold = ['550']
fire_escape = ['181', '512', '513', '514', '515', '516', '517', '518', '519', '520', '539', '812', '813', '814', 
               '815', '817', '818', '819', '820', '821']
heat =['226', '664', '665', '666', '668', '675', '964', '965', '975']
```

```python
def add_column(df, label, label_name):
    df.loc[df['OrderNumber'].isin(label), 'Violation Type'] = label_name
    return df
```

```python
add_column(new_df,lead_paint,'Lead Paint')
add_column(new_df,mold,'Mold')
add_column(new_df,fire_escape, 'Fire Escape')
add_column(new_df, heat, 'Heating')
```

```python
new_df['Violation Type'].unique()
```

```python
new_df.to_csv('Cleaned_Housing_Maintenance_Code_Violations.csv', index=False)
```
